```
# importing required modules 
import matplotlib.pyplot as plt 

# setting a style to use 
plt.style.use('ggplot') 

# create a figure 
fig = plt.figure(figsize=(8,8)) 

# define subplots and their positions in figure 

#TIMING DIAGRAM OF A 
# x axis values 
x = [0,80] 
# corresponding y axis values 
y = [1,1] 
#plt.ylim(0,10.5) 
plt1 = fig.add_subplot(321) 
plt1.plot(x, y, color ='r') 
plt1.set_title('Timing Diagram for $A$') 
plt.xlabel('Time') 
plt.ylabel('A') 
plt.ylim(-0.1,5.5) # change y axis limit
plt.xlim(-0.1,80) # change x axis limit

#TIMING DIAGRAM OF B
# x axis values 
x = [0,80] 
# corresponding y axis values 
y = [0,0] 
#plt.ylim(0,10.5) 
plt2 = fig.add_subplot(322) 
plt2.plot(x, y, color ='b') 
plt2.set_title('Timing Diagram for $B$') 
plt.xlabel('Time') 
plt.ylabel('B') 
plt.ylim(-0.1,5.5) 
plt.xlim(-0.1,80) 

#TIMING DIAGRAM OF C
# x axis values 
x = [0,80] 
# corresponding y axis values 
y = [1,1] 
plt3 = fig.add_subplot(323) 
plt3.plot(x, y, color ='g') 
x = [4,4]
y = [0,5]
plt3.set_title('Timing Diagram for $C$') 
plt.xlabel('Time') 
plt.ylabel('C') 
plt.ylim(-0.1,5.5) 
plt.xlim(-0.1,80) 

#TIMING DIAGRAM OF B'
# x axis values 
x = [0,20,20,40,80] 
# corresponding y axis values 
y = [0,0,1,1,1] 
#plt.ylim(0,10.5) 
plt4 = fig.add_subplot(324) 
plt4.plot(x, y, color ='r') 
plt4.set_title('Timing Diagram for $\overline{B}$') 
plt.xlabel('Time') 
plt.ylabel('$\overline{B}$') 
plt.ylim(-0.1,5.5)
plt.xlim(-0.1,80)

#TIMING DIAGRAM OF X
# x axis values 
x = [0,20,40,40,60,80] 
# corresponding y axis values 
y = [0,0,0,1,1,1] 
#plt.ylim(0,10.5) 
plt5 = fig.add_subplot(325) 
plt5.plot(x, y, color ='r') 
plt5.set_title('Timing Diagram for $X$') 
plt.xlabel('Time') 
plt.ylabel('X') 
plt.ylim(-0.1,5.5)
plt.xlim(-0.1,80)


#TIMING DIAGRAM OF Z
# x axis values 
x = [0,20,20,40,60,60,80] 
# corresponding y axis values 
y = [0,0,1,1,1,0,0] 
#plt.ylim(0,10.5) 
plt6 = fig.add_subplot(326) 
plt6.plot(x, y, color ='r') 
plt6.set_title('Timing Diagram for $Z$') 
plt.xlabel('Time') 
plt.ylabel('Z') 
plt.ylim(-0.1,5.5)
plt.xlim(-0.1,80)


fig.subplots_adjust(hspace=.5)
plt.savefig('graph.pdf')
# function to show the plot 
plt.show()
```
